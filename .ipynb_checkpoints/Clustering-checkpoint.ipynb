This notebook tries to address a very interesting application of KMeans Clustering to finding the optimal locations of fire and rescue stations in Mongomery County, MD. All 9-1-1 calls made from within the boundaries of the county have been used to asses the traffic density of all of these calls. All of these call vary a lot in their nature. Refer to the notebook '911_calls.ipynb' for a detailed analysis of all of these calls.

### Steps involved briefly

Step 1. The addresses of all existing fire and rescue stations was obtained from https://www.montgomerycountymd.gov/mcfrs/. Each one of this address was mapped with the corresponding latitudes and longitudes using a Google Maps API and saved in a .csv file.

Step 2. Traffic density 300,000 phone calls was analyzed to reveal existence of specific clusters.

```python
## Categorizing data by emergency type
df_title_cat = pd.DataFrame(df.type.str.split(':',1).tolist(),columns = ['emergency_type','emergency_subtype'])
df_cat = df.copy()
del df_cat['type']
df_cat.insert(4,'emergency_type',df_title_cat['emergency_type'].values)
df_cat.insert(5,'emergency_subtype',df_title_cat['emergency_subtype'].values)
```

## Clustering

```python
plt.plot(df_cat.latitude[(df_cat.latitude>39.8) & (df_cat.latitude<40.8)],df_cat.longitude[(df_cat.latitude>39.8) & (df_cat.latitude<40.8)],'o')
plt.show()
```

```python
stations_file = open('Fire_and_Rescue_Stations.csv','r').read()
stations_file = stations_file.split('\n')
```

```python
location_data = []
for element in stations_file:
    if len(element)>0:
        location_data.append(element.split(',')[1:])
```

```python
location_data = np.array(location_data).astype(np.float)
```

```python
plt.figure(figsize=(10,10))
plt.plot(location_data[:,1],location_data[:,0],'*',markersize=10)
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
only_ems = df_cat[df_cat.emergency_type=='EMS']
unscaled_ems = pd.DataFrame()
unscaled_ems['latitude'] = only_ems.latitude[(only_ems.latitude>39.8) & (only_ems.latitude<40.8)]
unscaled_ems['longitude'] = only_ems.longitude[(only_ems.latitude>39.8) & (only_ems.latitude<40.8)]
```

```python
track_location = pd.DataFrame()
max_lat = max(unscaled_ems['latitude'])
min_lat = min(unscaled_ems['latitude'])
max_long = max(unscaled_ems['longitude'])
min_long = min(unscaled_ems['longitude'])

track_location['scaled_lat'] = (unscaled_ems['latitude']-min_lat)/(max_lat-min_lat)
track_location['scaled_long'] = (unscaled_ems['longitude']-min_long)/(max_long-min_long)
```

```python
track_location.head()
```

```python
from sklearn.cluster import KMeans
n_clusters=97
clusters = KMeans(n_clusters=n_clusters)
clusters.fit(track_location.values)
```

```python
track_location['cluster'] = clusters.labels_
unscaled_cluster_centers = np.empty(clusters.cluster_centers_.shape)
unscaled_cluster_centers[:,0] = clusters.cluster_centers_[:,0]*(max_lat-min_lat)+min_lat
unscaled_cluster_centers[:,1] = clusters.cluster_centers_[:,1]*(max_long-min_long)+min_long
```

```python
import matplotlib
matplotlib.rcParams['agg.path.chunksize'] = 10000
matplotlib.rcParams['axes.prop_cycle']
plt.figure(figsize=(10,10))
for i in range(n_clusters):
    plt.plot(unscaled_ems['longitude'][track_location.cluster==i],unscaled_ems['latitude'][track_location.cluster==i],'o')
plt.plot(unscaled_cluster_centers[:,1],unscaled_cluster_centers[:,0],'g*',markersize=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.axis('equal')
plt.show()
plt.close()
```

```python
clusters.inertia_
```

```python
plt.figure(figsize=(10,10))
plt.plot(location_data[:,1],location_data[:,0],'*',markersize=10,label='Current locations')
plt.plot(unscaled_cluster_centers[:,1],unscaled_cluster_centers[:,0],'*',markersize=10,label='Optimal locations')
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

```python
def calc_inertia(unscaled_ems,location_data):
    '''
    unscaled_ems: Dataframe of all location values
    location_data: Array of locations of all stations
    '''
    inertia = 0
    
    for loc in range(unscaled_ems.shape[0]):
        x,y=unscaled_ems.values[loc,0],unscaled_ems.values[loc,1]
        dist=[]
        for station in range(location_data.shape[0]):
            xs,ys = location_data[station,0],location_data[station,1]
            dist.append(np.sqrt((x-xs)**2+(y-ys)**2))
        inertia += min(dist)
        
    return inertia
```

```python
inertia_original = calc_inertia(unscaled_ems,location_data)
```

```python
inertia_calc = calc_inertia(unscaled_ems,unscaled_cluster_centers)
```
