This notebook tries to address a very interesting application of K-Means Clustering to finding the optimal locations of fire and rescue stations in Mongomery County, MD. All 9-1-1 calls made from within the boundaries of the county have been used to asses the traffic density of all of these calls. All of these call vary a lot in their nature. Refer to the notebook '911_calls.ipynb' for a detailed analysis of all of these calls.

### Steps involved briefly

Step 1. The addresses of all existing fire and rescue stations was obtained from https://www.montgomerycountymd.gov/mcfrs/. Each one of this address was mapped with the corresponding latitudes and longitudes using a Google Maps API and saved in a .csv file.

Step 2. Traffic density of 300,000 phone calls was analyzed to reveal existence of specific clusters. The cluster locations were optimized using the K-Means algorithm.

Step 3. Comparison of newly obtained locations with the original locations reveal a clear improvement.

```python
# Import required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 10
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('911 calls- Mongomery county/911.csv')
df = df[df['zip'] > 10000]
```

```python
# lat: Latitude
# lng: Longitude
# desc: Description of Emergency
# zip: ZIP Code
# title: Title of Emergency
# time: StampDate and time of the call
# twp: Town
# addr: Address
```

```python
# First 5 rows
(df.head())
```

```python
# Last 5 rows
df.tail()
```

```python
df.columns = ['latitude','longitude','description','zip','type','time',
              'town','address','if_emergency']
```

```python
df.dtypes
```

```python
df['time'] = pd.to_datetime(df['time'])

print ('The first phone call in the dataset was made on:',df['time'].max())
print ('The last phone call in the dataset was made on:',df['time'].min())

print ('The dataset spans a period of:',df['time'].max() - df['time'].min())
```

```python
## Categorizing data by emergency type

df_title_cat = pd.DataFrame(df.type.str.split(':',1).tolist(),
                            columns = ['emergency_type','emergency_subtype'])
df_cat = df.copy()
del df_cat['type']
df_cat.insert(4,'emergency_type',df_title_cat['emergency_type'].values)
df_cat.insert(5,'emergency_subtype',df_title_cat['emergency_subtype'].values)
```

## Clustering

```python
# See where all the calls originated from

plt.plot(df_cat.latitude[(df_cat.latitude>39.8) & 
                         (df_cat.latitude<40.8)],df_cat.longitude[(df_cat.latitude>39.8) 
                                                                  & (df_cat.latitude<40.8)],'o')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.title('All Calls')
plt.show()
```

```python
# Read locations document

stations_file = open('Fire_and_Rescue_Stations.csv','r', encoding="latin-1").read()
stations_file = stations_file.split('\n')
```

```python
location_data = np.array([element.split(',')[1:] for element in stations_file 
                          if len(element) > 0]).astype(np.float)
```

```python
plt.figure(figsize=(10,10))
plt.plot(location_data[:,1],location_data[:,0],'*',markersize=10)
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Existing Locations of all Fire and Rescue Stations')
plt.show()
```

```python
# Query only the EMS emergencies
only_ems = df_cat[df_cat.emergency_type=='EMS']
unscaled_ems = pd.DataFrame()
unscaled_ems['latitude'] = only_ems.latitude[(only_ems.latitude>39.8) & (only_ems.latitude<40.8)]
unscaled_ems['longitude'] = only_ems.longitude[(only_ems.latitude>39.8) & (only_ems.latitude<40.8)]
```

```python
# Store all emergency locations in a dataframe
track_location = pd.DataFrame()
max_lat = max(unscaled_ems['latitude'])
min_lat = min(unscaled_ems['latitude'])
max_long = max(unscaled_ems['longitude'])
min_long = min(unscaled_ems['longitude'])

track_location['scaled_lat'] = (unscaled_ems['latitude']-min_lat)/(max_lat-min_lat)
track_location['scaled_long'] = (unscaled_ems['longitude']-min_long)/(max_long-min_long)
```

```python
track_location.head()
```

### Build a Model Using the K-Means Algorithm

```python
n_clusters=97
clusters = KMeans(n_clusters=n_clusters)
clusters.fit(track_location.values)
```

```python
# All the new cluster centers to the original dataframe
track_location['cluster'] = clusters.labels_
unscaled_cluster_centers = np.empty(clusters.cluster_centers_.shape)
unscaled_cluster_centers[:,0] = clusters.cluster_centers_[:,0]*(max_lat-min_lat)+min_lat
unscaled_cluster_centers[:,1] = clusters.cluster_centers_[:,1]*(max_long-min_long)+min_long
```

```python
import matplotlib
matplotlib.rcParams['agg.path.chunksize'] = 10000
matplotlib.rcParams['axes.prop_cycle']

plt.figure(figsize=(10,10))

for i in range(n_clusters):
    plt.plot(unscaled_ems['longitude'][track_location.cluster==i],unscaled_ems['latitude'][track_location.cluster==i],'o')
    
plt.plot(unscaled_cluster_centers[:,1],unscaled_cluster_centers[:,0],'g*',markersize=10)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Generated Clusters and Corresponding Cluster Centers')

plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(location_data[:,1],location_data[:,0],'*',markersize=10,label='Current locations')
plt.plot(unscaled_cluster_centers[:,1],unscaled_cluster_centers[:,0],'*',markersize=10,label='Optimal locations')
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Comparing New Locations with the Original Locations')
plt.legend()
plt.show()
```

```python
def calc_inertia(unscaled_ems,location_data):
    '''
    unscaled_ems: Dataframe of all location values
    location_data: Array of locations of all stations
    '''
    inertia = 0
    
    for loc in range(unscaled_ems.shape[0]):
        x,y=unscaled_ems.values[loc,0],unscaled_ems.values[loc,1]
        dist=[]
        for station in range(location_data.shape[0]):
            xs,ys = location_data[station,0],location_data[station,1]
            dist.append(np.sqrt((x-xs)**2+(y-ys)**2))
        inertia += min(dist)
        
    return inertia
```

```python
inertia_original = calc_inertia(unscaled_ems,location_data)
inertia_calc = calc_inertia(unscaled_ems,unscaled_cluster_centers)
```

```python
print ('Original inertia:',inertia_original)
print ('New inertia:',inertia_calc)
```

```python
# Total travel distance saved
# 1 degree = 69.172 miles

print ('Total travelling distance saved:',int((inertia_original - inertia_calc)*69.172),'miles.')

# Average per trip distance saved
print ('Average distance saved per trip:',np.round((inertia_original - inertia_calc)*69.172/len(df),4),'miles.')

# Time saved
print ('Time saved:',np.round((inertia_original - inertia_calc)*69.172/len(df)/60,4),'hours')
print ('Time saved:',np.round((inertia_original - inertia_calc)*69.172/len(df),4),'min.')
print ('Time saved:',np.round((inertia_original - inertia_calc)*69.172/len(df)*60,4),'s.')
```
